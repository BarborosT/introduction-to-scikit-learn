```python
favorite_food_prices ={"Almond butter":23,
                      "Blueberries":12,
                      "Eggs":7}
```

```python
fig, ax=plt.subplots()
ax.barh(list(favorite_food_prices.keys()),
       list(favorite_food_prices.values()))
ax.set(title="Danial's favorite foods",
      xlabel="Food",
      ylabel="Price($)");
```

```python
import pandas as pd
```

```python
car_sales = pd.read_csv("car-sales-missing-data.csv")
car_sales
```

```python
car_sales["Doors"].plot(); #aslında burda PRİCE olması lazım ama kabul etmiyor
```

```python
# Remove the symbol, the final two numbers from the price column and converst iit to numbers
    
car_sales["Price"] =car_sales["Price"].str.replace("[\$\,\.]", "")
car_sales["Price"] = car_sales["Price"].str[:-2]
```

```python
car_sales["Total Sales"] = car_sales["Price"].astype(int).cumsum()
```

```python
car_sales["Sale Date"] = pd.date_range("13/1/2020", periods=len(car_sales))
```

```python
car_sales.plot(x="Odometer", y="Price", kind="scatter")
```

```python
over_50 =heart_disease[heart_disease["age"]>50]
fig,ax =plt.subplots(figsize=(10,6))
scatter = ax.scatter(over_50["age"],
                    over_50["chol"],
                    c=over_50["target"])
ax.set(title="Heart-disase-and-max-heart-rate-levels",
      xlabel="Age",
      ylabel="Cholesterol");
ax.legend(*scatter.legend_elements(),title="Target")
ax.axhline(over_50["chol"].mean(),
          linestyle="--");
```
